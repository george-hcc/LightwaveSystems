# Exercício 02

Determine a transformada de Fourier discreta do pulso triangular:

```python
def triang(t):
    return (1-abs(t)) * (abs(t)<1).astype(float)
```

e, em seguida, do sinal periódico triangular:

```python
def periodic_triang(t):
    return triang(num.mod(t+1, 2)-1)
```

```python
Fs = 10000
t = num.arange(-10, 10, 1/Fs)
f = num.arange(-3, 3, 1/20.)

plt.subplot(121)
plt.plot(t, triang(t))
plt.subplot(122)
plt.plot(t, periodic_triang(t));
```
