# Atividade 03 - Banda passante e mais FFT

```python
%matplotlib inline
import numpy as num
import matplotlib.pyplot as plt
from scipy import signal
```

```python
def rect(t):
    return (abs(t)<0.5).astype(float)

def triang(t):
    return (1-abs(t)) * (abs(t)<1).astype(float)

def periodic_triang(t):
    return triang(num.mod(t+1, 2)-1)

window = lambda t: rect((t-1)/4)

def windowed_periodic_triang(t):
    return periodic_triang(t) * window(t)
```

```python
Fs = 10000
t = num.arange(-5, 5, 1/Fs)
```

### 1. Determine a transformada de Fourier do trem de pulsos triangulares ilustrado abaixo. Quais as frequências das componentes de frequência diferentes da frequência DC?

```python
plt.figure(figsize=(12,6))
plt.plot(t, periodic_triang(t), 'g-x', label='periodic', markevery=2000);
plt.grid();
plt.xlabel("t", fontsize = 14);
plt.xticks(fontsize=14);
plt.xlim(-5, 5);
```

### 2.Determine a transformada de Fourier do sinal ilustrado abaixo. Sobreponha a transformada de Fourier deste item e do item anterior. Qual a diferença entre estas transformadas? Qual o significado desta transformação?

```python
plt.figure(figsize=(12,6))
plt.plot(t, windowed_periodic_triang(t), 'b-', lw=2);
plt.grid();
plt.xlabel("t", fontsize = 14);
plt.xticks(fontsize=14);
plt.xlim(-5, 5);
```
