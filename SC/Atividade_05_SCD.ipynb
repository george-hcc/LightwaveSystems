# Atividade 05

Considere dois streams de bits $x_1(t)$ e $x_2(t)$:

```python
N  = 2**12; # número de bits
Fb = 10;    # bit rate   [bps]
Tb = 1/Fb;  # bit period [s]
Ns = 16;    # samples per bit
Fs = Fb*Ns  # sample rate
```

```python
NFFT  = N * Ns/4;
t = num.arange(0, N/Fb, 1/Fs);
x1, b1, d1 = ss.NRZ_bits(N, Ns, 'rect');
x2, b2, d2 = ss.NRZ_bits(N, Ns, 'rect');
```

```python
plt.figure(figsize=(12,6));
plt.subplot(121);
plt.plot(t[0:16*Ns:], x1[0:16*Ns:],
         t[0:16*Ns:Ns], 2*d1[0:16:]-1, 'ro');
plt.grid();
plt.xlabel(r'Tempo [s]', fontsize = 14);
plt.ylabel(r'Amplitude x1[n]', fontsize = 14);
plt.title(r'Número total de amostras: ' + repr(N), fontsize = 14);
plt.subplot(122);
plt.plot(t[0:16*Ns:], x2[0:16*Ns:],
         t[0:16*Ns:Ns], 2*d2[0:16:]-1, 'ro');
plt.grid();
plt.xlabel(r'Tempo [s]', fontsize = 14);
plt.ylabel(r'Amplitude x2[n]', fontsize = 14);
plt.title(r'Número total de amostras: ' + repr(N), fontsize = 14);
plt.tight_layout();
```

```python
f1, X1 = ss.ft_approx(x1, t, NFFT);
f2, X2 = ss.ft_approx(x2, t, NFFT);
```

```python
plt.figure(figsize=(12,6));
plt.subplot(121);
plt.plot(f1, 10*num.log10(num.abs(X1)));
plt.grid()
plt.xlim([-21,21])
plt.ylim([-30,20])
plt.title(r'Magnitude espectral', fontsize = 14);
plt.xlabel(r'Frequência (Hz)', fontsize = 14);
plt.ylabel(r'$10log_{10}|X1(f)|$', fontsize = 14);

plt.subplot(122);
plt.plot(f2, 10*num.log10(num.abs(X2)));
plt.grid()
#plt.xlim([-21,21])
plt.ylim([-30,20])
plt.title(r'Magnitude espectral', fontsize = 14);
plt.xlabel(r'Frequência (Hz)', fontsize = 14);
plt.ylabel(r'$10log_{10}|X2(f)|$', fontsize = 14);
```
