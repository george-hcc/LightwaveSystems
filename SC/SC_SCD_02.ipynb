# Sistemas digitais coerentes 

```python
import numpy as num
import matplotlib.pyplot as plt
import sk_dsp_comm.sigsys as ss
from scipy import signal
```

## Stream de bits

```python
N  = 2**6; # número de bits
Fb = 10;    # bit rate   [bps]
Tb = 1/Fb;  # bit period [s]
Ns = 16;    # samples per bit
Fs = Fb*Ns  # sample rate
```

```python
x1, b1, d1 = ss.NRZ_bits(N, Ns, 'rect');
x2, b2, d2 = ss.NRZ_bits(N, Ns, 'rect');
t = num.arange(0, N/Fb, 1/Fs);
NFFT  = N * Ns/2;
f, X1 = ss.ft_approx(x1, t, NFFT);
f, X2 = ss.ft_approx(x2, t, NFFT);
```

```python
plt.figure(figsize=(12,6));
plt.subplot(121);
plt.plot(t, x1,
         t[0::Ns], 2*d1-1, 'ro');
plt.grid();
plt.xlabel(r'Tempo [s]', fontsize = 14);
plt.ylabel(r'Amplitude x[n]', fontsize = 14);
plt.title(r'Número total de amostras: ' + repr(N), fontsize = 14);
plt.subplot(122);
plt.plot(t, x2,
         t[0::Ns], 2*d2-1, 'ro');
plt.grid();
plt.xlabel(r'Tempo [s]', fontsize = 14);
plt.ylabel(r'Amplitude x[n]', fontsize = 14);
plt.title(r'Número total de amostras: ' + repr(N), fontsize = 14);
plt.tight_layout();
```

```python
f, X1 = ss.ft_approx(x1, t, NFFT);
f, X2 = ss.ft_approx(x2, t, NFFT);
```

```python
plt.figure(figsize=(12,6));
plt.subplot(121);
plt.plot(f, 10*num.log10(num.abs(X1)));
plt.grid()
plt.xlim([-21,21])
plt.ylim([-30,20])
plt.title(r'Magnitude espectral', fontsize = 14);
plt.xlabel(r'Frequência (Hz)', fontsize = 14);
plt.ylabel(r'$10log_{10}|X(f)|$', fontsize = 14);

plt.subplot(122);
plt.plot(f, 10*num.log10(num.abs(X2)), 'r.-');
plt.grid()
plt.xlim([-21,21])
plt.ylim([-30,20])
plt.title(r'Magnitude espectral', fontsize = 14);
plt.xlabel(r'Frequência (Hz)', fontsize = 14);
plt.ylabel(r'$10log_{10}|X(f)|$', fontsize = 14);
```

```python
Fc = 10; #10 Hz
x1_tx = x1 * num.cos(2 * num.pi * Fc * t);
x2_tx = x2 * num.cos(2 * num.pi * Fc * t);
f, X1_tx = ss.ft_approx(x1_tx, t, NFFT);
f, X2_tx = ss.ft_approx(x2_tx, t, NFFT);
```

```python
plt.figure(figsize=(12,6));
plt.subplot(121);
plt.plot(f, 10*num.log10(num.abs(X1_tx)));
plt.grid()
plt.xlim([-50,50])
plt.ylim([-30,20])
plt.title(r'Magnitude espectral', fontsize = 14);
plt.xlabel(r'Frequência (Hz)', fontsize = 14);
plt.ylabel(r'$10log_{10}|X(f)|$', fontsize = 14);

plt.subplot(122);
plt.plot(f, 10*num.log10(num.abs(X2_tx)), 'r.-');
plt.grid()
plt.xlim([-50,50])
plt.ylim([-30,20])
plt.title(r'Magnitude espectral', fontsize = 14);
plt.xlabel(r'Frequência (Hz)', fontsize = 14);
plt.ylabel(r'$10log_{10}|X(f)|$', fontsize = 14);
```

```python
Fc1 = 10; #10 Hz
Fc2 = 20; #10 Hz
x1_tx = x1 * num.cos(2 * num.pi * Fc1 * t);
x2_tx = x2 * num.cos(2 * num.pi * Fc2 * t);
f, X1_tx = ss.ft_approx(x1_tx, t, NFFT);
f, X2_tx = ss.ft_approx(x2_tx, t, NFFT);
```

```python
plt.figure(figsize=(12,6));
plt.subplot(121);
plt.plot(f, 10*num.log10(num.abs(X1_tx)));
plt.grid()
plt.xlim([-50,50])
plt.ylim([-30,20])
plt.title(r'Magnitude espectral - Fc = 10 Hz', fontsize = 14);
plt.xlabel(r'Frequência (Hz)', fontsize = 14);
plt.ylabel(r'$10log_{10}|X(f)|$', fontsize = 14);

plt.subplot(122);
plt.plot(f, 10*num.log10(num.abs(X2_tx)), 'r.-');
plt.grid()
plt.xlim([-50,50])
plt.ylim([-30,20])
plt.title(r'Magnitude espectral - Fc = 20 Hz', fontsize = 14);
plt.xlabel(r'Frequência (Hz)', fontsize = 14);
plt.ylabel(r'$10log_{10}|X(f)|$', fontsize = 14);
```

```python
plt.figure(figsize=(12,6));
plt.plot(f, 10*num.log10(num.abs(X1_tx)), label = '$X1_{tx}$');
plt.plot(f, 10*num.log10(num.abs(X2_tx)), label = '$X2_{tx}$');
plt.legend(loc='upper left');
plt.grid()
plt.xlim([-50,50])
plt.ylim([-30,10])
plt.title(r'Magnitude espectral', fontsize = 14);
plt.xlabel(r'Frequência (Hz)', fontsize = 14);
plt.ylabel(r'$10log_{10}|X(f)|$', fontsize = 14);
```

```python
x_tx = x1_tx + x2_tx;
f, X_tx = ss.ft_approx(x_tx, t, NFFT);
```

```python
plt.figure(figsize=(12,6));
plt.plot(f, 10*num.log10(num.abs(X_tx)));
plt.grid()
plt.xlim([-50,50])
plt.ylim([-30,20])
plt.title(r'Magnitude espectral - $X1_{tx} + X2_{tx}$', fontsize = 14);
plt.xlabel(r'Frequência (Hz)', fontsize = 14);
plt.ylabel(r'$10log_{10}|X(f)|$', fontsize = 14);
```

## Recuperar o sinal modulado em 10 Hz.

```python
# Demodulação
x_rx = x_tx * num.cos(2 * num.pi * Fc1 * t);
f, X1_rx = ss.ft_approx(x_rx, t, NFFT);
# Filtro
w = Fc1*num.sinc(Fc1*t);
f,W = ss.ft_approx(w, t, NFFT);
# Convolução
x_rx_c = num.convolve(x_rx, w);
f, Xrxc = ss.ft_approx(x_rx_c, t, NFFT);
```

```python
plt.figure(figsize=(12,6));
plt.plot(f, 10*num.log10(num.abs(X1_rx)), label = '$X1_{rx}$');
plt.plot(f, 10*num.log10(num.abs(W)), label = '$W$');
plt.plot(f, 10*num.log10(num.abs(X1_rx*W)), label = '$W * X1_{rx}$');
plt.plot(f, 10*num.log10(num.abs(X1)), label = '$X1$');
plt.plot(f, 10*num.log10(num.abs(Xrxc)), label = '$convolve(x1_{rx}, w)$');
plt.legend(loc='upper left');
plt.grid()
plt.xlim([-20,20])
plt.ylim([-30,25])
plt.title(r'Magnitude espectral', fontsize = 14);
plt.xlabel(r'Frequência (Hz)', fontsize = 14);
plt.ylabel(r'$10log_{10}|X(f)|$', fontsize = 14);
```

```python
plt.figure(figsize=(12,6));
plt.plot(t, x_rx_c[0:int(len(t)):],
         t[0::Ns], 2*d1-1, 'ro');
plt.grid();
plt.xlabel(r'Tempo [s]', fontsize = 14);
plt.ylabel(r'Amplitude x[n]', fontsize = 14);
plt.title(r'Número total de amostras: ' + repr(N), fontsize = 14);
```

```python
x_rx = num.sign(x_rx_c[0:int(len(t)):]);
```

```python
plt.figure(figsize=(12,6));
plt.plot(t, x_rx);
plt.plot(t[0::Ns], x_rx[0::Ns], 'bx');
plt.plot(t[0::Ns], 2*d1-1, 'r+');
plt.grid();
plt.xlabel(r'Tempo [s]', fontsize = 14);
plt.ylabel(r'Amplitude x[n]', fontsize = 14);
plt.title(r'Número total de amostras: ' + repr(N), fontsize = 14);
```
